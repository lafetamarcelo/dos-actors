```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
plt.style.use("dark_background")
```

```python
data = pd.read_parquet("m2_logs.parquet")
data.columns
```

```python
t = np.arange(len(data))/8e3
fig, ax = plt.subplots()
ax.plot(t, np.vstack(data["MCM2Lcl6D"])[:,:6])
ax.grid()
ax.set_xlabel("Time[s]")
ax.set_ylabel("MCM2Lcl6D")
#ax.set_xlim([0,1e-1])
```

```python
data = pd.read_parquet("data.parquet")
data.columns
```

```python
t = np.arange(len(data))/1e3
fig, axs = plt.subplots(nrows=3, sharex=True, figsize=(15, 15))
ax = axs[0]
ax.plot(t, np.vstack(data["MountEncoders"])[:, :4].mean(1))
ax.grid()
ax.set_title("Azimuth")
ax = axs[1]
ax.plot(t, np.vstack(data["MountEncoders"])[:, 4:10].mean(1))
ax.grid()
ax.set_title("Elevation")
ax = axs[2]
ax.plot(t, np.vstack(data["MountEncoders"])[:, 10:].mean(1))
ax.grid()
ax.set_xlabel("Time [s]")
ax.set_title("GIR")
```
