```python
from scipy.special import kv
def variance(a,b,c):
    return np.pi * a * b**(-2. * (c - 1.)) / (c - 1.)
def covariance(a,b,c, r):
    if r == 0:
        return variance(a,b,c)
    else:
        red = 2. * np.pi * b * r
        return (2. * np.pi * a / gamma(c)) \
        * (b**(-2. * (c - 1.)) / 2**(c - 1.)) \
        * red**(c - 1.) \
        * kv(c - 1., red)
def structure_function( a,b,c, r):
    return 2. * (variance(a,b,c) - covariance(a,b,c,r))


def first_order_structure_function(a, c, r):
    denom = gamma(c)**(2) * np.sin(np.pi * c)
    num = -2 * np.pi**(2) * a
    return (num / denom) * (np.pi * r)**(2. * (c - 1.))

```

```python
def otfad(x,id):
    otf = 0
    for i in range(1000):
        (g,q) = gq(id,i)
        otf += np.exp(-0.5*6.88*(x/r0)**(5/3)) * np.exp(-0.5*g*x**q)
    return otf/1000


def otfd(x, id):
    otf = 0
    for i in range(1000):
        (g, q) = gq(id, i)
        otf += np.exp(-0.5*g*x**q)
    return otf/1000

def dfun(x,id): return x * otfad(x,id)**2
```
