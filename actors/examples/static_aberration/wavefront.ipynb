```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
plt.style.use("dark_background")
```

```python
def load(name):
    data = pd.read_parquet(name+".parquet")
    return np.vstack(data["Wavefront"]).reshape(769, -1)
```

```python
plt.imshow(load("raw-polishing_769"))
plt.colorbar()
```
