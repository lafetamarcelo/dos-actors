```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
plt.style.use('dark_background')
```

```python
def fm(key): 
    fm = np.vstack(data[key]).reshape(len(data), -1, 6)
    return np.split(fm,2,axis=2)
```

```python
def mag(x):
    return np.sqrt(np.sum(x**2,axis=2))
```

```python

data = pd.read_parquet("data.parquet")
print(data.columns)
```

```python
[f, m] = fm("CFD2021106F")
fig,axs = plt.subplots(nrows=2,sharex=True,figsize=(10,10))
ax = axs[0]
ax.plot(mag(f).sum(1))
ax.grid()
ax = axs[1]
ax.plot(mag(m).sum(1))
ax.grid()
```
